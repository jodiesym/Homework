```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os

# Import API key
from api_keys import g_key
```

```python
city_file = "./cities.csv"
```

```python
city_df = pd.read_csv(city_file)
city_df
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = city_df[["lat", "lon"]].astype(float)
# Convert Poverty Rate to float and store
humidity = city_df["humidity"].astype(float)
```

```python
# Create a humidity Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 2)

fig.add_layer(heat_layer)
fig
```

```python
# now we need to drop the city data to get the ideal vacation places: 
# Temperature between 21(c) to 26(c)
# wind speed less than 6 mph
# Zero Cloudiness

vacation_df = city_df[city_df['cloudiness']==0]
vacation_df = vacation_df[vacation_df['wind_speed'] <= 6]
vacation_df = vacation_df[vacation_df['max_temp'] <= 26]
vacation_df = vacation_df[vacation_df['max_temp'] >= 21]
vacation_df.count()
```

```python
vacation_df
```

```python
# drop useless info in hotel df
hotel_df = vacation_df.drop(columns = ['cloudiness', 'date','humidity','main_temp','max_temp','pressure','visibility','wind.deg',
                                      'wind_speed','timezone'])
# add hotel name column to the df
hotel_df['hotel name']=""
hotel_df
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
# for loop to look through all locations in the Hotel data, and fill in the first found hotel name
```

```python
# go through the hotel_df and set parameters per latitude and longitude
for index, row in hotel_df.iterrows():
    lat = row['lat']
    lon = row['lon']
    params = {
    "location": f"{lat},{lon}",
    "rankby": "distance",
    "type": "hotel",
    "key": g_key}
    
    restr_type = row['city']
    
    # print out log of each city as it's being processed
    
    print(f"Retrieving Results for Index {index}: {restr_type}.")
    response = requests.get(base_url, params=params).json()
    
    results = response['results']
    
    try:
        print(f"Closest {restr_type} hotel is {results[0]['name']}.")
        
        # Save the first search result in my hotel_df 
        hotel_df.loc[index,'hotel name'] = results[0]['name']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
    
    
    
```

```python
# print out the df to make sure I have everything
hotel_df
```

```python
# now add markers layer on top of the heat map: 
```

```python
marker_locations = hotel_df[['lat', 'lon']]

# make lists, then zip all lists together --> then make a dictionary
cities = hotel_df["city"].tolist()
countries = hotel_df["country"].tolist()
hotels = hotel_df["hotel name"].tolist()
latitudes = hotel_df["lat"].tolist()

info_dict = dict((z[0],list(z[1:])) for z in zip(latitudes,cities,countries,hotels))

# Pin the google map and add all info by going through the dictionary I just made
markers = gmaps.marker_layer(marker_locations,
    info_box_content= [f"City: {info_dict[info][0]},Country:{info_dict[info][1]}, Name: {info_dict[info][2]}" for info in info_dict] )

fig.add_layer(markers)
fig
```
